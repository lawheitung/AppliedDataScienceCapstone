```python
# Get postal code from Wikipedia
url = "https://www.irishtourist.com/tourist-information/getting-to-and-around-ireland/maps-of-ireland/dublin-postal-code-map/"
request = requests.get(url).text

# Passing the html to bs4
soup = BeautifulSoup(request, 'html.parser')

print(soup.prettify())
```

```python
info = []
show = []
for i in range(4,4+22):
    table = soup.select("p:nth-of-type({})".format(i))
    details = table[0].find_all('br')
    #print(details)
    area = []
    for j in details:
        postal_code = []
        
        postal_code.append(details[0].previous_sibling)
        nghborhd = j.next_sibling
        postal_code.append(nghborhd)
        area.append(postal_code)
    del area[0]
    
    print(area)
    info.extend(area)

#for k in info:
 #   for l in k:
  #      l.lstrip()
   #     print(l)



#print(postal_code)


```
